```python
transcriptdf = pd.read_csv('Transcripts_df.csv')
```

```python
transcriptdf.head(5)
```

```python
#look at the transcript column

transcriptdf['Transcript']
```

```python
#create new column to see if I can remove punctuation
transcriptdf['words'] =  transcriptdf['Transcript']
```

```python
transcriptdf['words']
```

```python
#removes punctuation
transcriptdf['words'] = transcriptdf['words'].str.replace(r'[^\w\s]+', ' ')

transcriptdf['words'] 
```

```python
#word frequencies of all words using words column
wordoccur = transcriptdf['words'].str.split(expand=True).stack().value_counts()
wordoccur.head(50)
```

```python
#word frequencies of all words using Transcripts column
wordoccur2 = transcriptdf['Transcript'].str.split(expand=True).stack().value_counts()
wordoccur2.head(50)
#will still need to parse out the all capitals speakers names in the transcript column
```
